```python
from arch.univariate.base import ARCHModelForecastSimulation
from arch import arch_model
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics
from statistics import stdev

data=pd.read_csv('/content/drive/MyDrive/Labs/TT/AMD.csv')
data=data.set_index('Date')
y=data[['Open']]

res = arch_model(y, p=1, o=0,q=1).fit()
fc=res.forecast(start=0,horizon=1)
sim_mod = arch_model(None, p=1, o=0,q=1)
sim_data = sim_mod.simulate(res.params, len(y)-1)

y1=np.array(y)
r=100 * y.pct_change().dropna()
# r=r.rolling(2).std()
print(r)

d=sim_data[['volatility']]
print(d)

rolling_predictions = []
test_size = 997

for i in range(test_size):
    train = r[:-(test_size-i)]
    model = arch_model(train, p=1,o=0,q=1)
    model_fit = model.fit(disp='off')
    pred = model_fit.forecast(horizon=1)
    rolling_predictions.append(np.sqrt(pred.variance.values[-1,:][0]))


rolling_predictions = pd.Series(rolling_predictions)
plt.plot(np.array(r)[:],label='calculated')
plt.plot(rolling_predictions[:],label='predict')
plt.legend()
plt.show()

print(metrics.mean_absolute_error(r[1:],d[1:]))
print(metrics.mean_squared_error(r[1:],d[1:]))
print(np.sqrt(metrics.mean_absolute_error(r[1:],d[1:])))

```
